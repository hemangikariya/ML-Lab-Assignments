# **Part C: Business Insights**

1. Which email provider has the highest number of users?

```python
df["Email Provider"]=df["Email"].apply(lambda x:x.split("@")[1])
```

```python
df["Email Provider"].value_counts().head()
```

**2. Calculate:**

○ Total sales | Average sales | Median purchase price

```python
# Total Sales
df["Purchase Price"].sum()
```

```python
# Average Sales
df["Purchase Price"].mean()
```

```python
# Median
df["Purchase Price"].median()
```

**3. Create a new column:**
Purchase Category based on Purchase Price:
Price   Category
< 30    Low
30–60   Medium
 > 60    High

```python
def category(price):
  if price<30:
    return"Low"
  elif price<=60:
    return"Medium"
  else:
    return "High"

  df["Purchase Category"]=df["Purchase Price"].apply(category)

# Using Lambda
# df["Purchase Category"]=df["Purchase Price"].apply(
# lambda x:"Low" if x<30 else ("Medium" if x<=60 else "High"))
```

4. Show the percentage of customers in each category.

```python
df["Purchase Category"].value_counts(normalize=True)*100
```

# **Part D: Advanced Pandas**

**Use:**

● groupby()
● sort_values()
● value_counts()
● apply()
● lambda
● isin()

1. Which company has the highest average purchase price?

```python
df.groupby("Company")["Purchase Price"].mean().sort_values(ascending=False).head(1)
```

2. Which credit card provider contributes the highest revenue?

```python
df.groupby("CC Provider")["Purchase Price"].sum().sort_values(ascending=False)
```

3. Which browser users spend the most?

```python
df.groupby("Browser")["Purchase Price"].mean().sort_values(ascending=False)
```
