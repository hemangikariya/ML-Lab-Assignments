# Step 4: Individual Descriptive Statistics

```python
print(f"Mean of weight:: {df['Weight'].mean()}")
print(f"Median of weight:: {df['Weight'].median()}")
print(f"Mode of Metcolor :: {df['MetColor'].mode()}")
print(f"Standard deviation of weight:: {df['Weight'].std()}")

# Calculate percentiles using NumPy
percentile_25 = np.percentile(df['Weight'], 25)
percentile_75 = np.percentile(df['Weight'], 75)
print(f"Percentile at 25% {percentile_25}")
print(f"Percentile at 75% {percentile_75}")
```

```python
# isnull -> Gives you how many null values are there in each column

df.isnull()

# False means not null
# True means null
```

```python
df.isnull().sum()
```

```python
# duplicated ->  It gives how many duplicate data are in dataset
df.duplicated()

# so, here false means no duplicate
```

```python
# let take sum of it and check is there any duplicate rows are there in dataset
int(df.duplicated().sum())

# So ans wer is 0 means no duplicate row
```
