# Data Cleaning
# Drop some rows having NA values

```python
# index_col = 0 make first column as index
# na_values Replace the values given in list as NA
cars_df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Toyota.csv",index_col=0,na_values=['??','????'])
cars_df.info()
cars_df.dropna(inplace=True)
```

# Change values of Doors (Three=3, Four=4, Five=5)

```python
cars_df['Doors'].unique()
```

```python
# cleaning column 'Doors'
cars_df.replace({'three':3},inplace=True)
#cars_df['Doors'].replace('three',3,inplace=True)
cars_df['Doors'].replace('four',4,inplace=True)
cars_df['Doors'].replace('five',5,inplace=True)
```

```python
cars_df['Doors'].unique()
```

```python
cars_df['Doors']=cars_df['Doors'].astype('int64')
cars_df.info()
```

# Two-way cross Tabulation
# A two-way cross-tabulation summarizes the association between two categorical (i.e., nominal, ordinal) variables.

```python
cars_df['FuelType'].value_counts()
```

```python
pd.crosstab(index=cars_df['FuelType'],columns='count')
```

```python
pd.crosstab(index=cars_df['Automatic'],columns=cars_df['FuelType'],dropna=True)
```

```python
pd.crosstab(index=cars_df['MetColor'],columns=cars_df['FuelType'],dropna=True)
```

```python
pd.crosstab(index=cars_df['Doors'],columns=cars_df['FuelType'],dropna=True)
```
