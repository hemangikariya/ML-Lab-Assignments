# Data Cleaning
# Drop some rows having NA values

```python
# index_col = 0 make first column as index
# na_values Replace the values given in list as NA
cars_df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Toyota.csv',index_col=0,na_values=['??','????'])
cars_df.info()
```

```python
cars_df
```

```python
cars_df.dropna(inplace=True)
cars_df.info()
```

```python
cars_df.describe()
```

# **Change values of Doors (Three=3, Four=4, Five=5)**

```python
cars_df['Doors'].unique()
```

```python
# cleaning column 'Doors'
cars_df['Doors'].replace({'three':3},inplace=True)
#cars_df['Doors'].replace('three',3,inplace=True)
cars_df['Doors'].replace('four',4,inplace=True)
cars_df['Doors'].replace('five',5,inplace=True)
```

```python
cars_df['Doors'].unique()
```

```python
# convert '3' in to number
cars_df['Doors']=cars_df['Doors'].astype('int64')
cars_df.info()
```

```python
cars_df['Doors'].unique()
```

# **Box Plot**

```python
#!pip install matplotlib
#!pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.boxplot(y=cars_df['Price'])
plt.title('Box plot of Price')
plt.ylabel('Price')
cars_df.describe()
```

```python
cars_df['Doors'].unique()
```

```python
sns.boxplot(y=cars_df['Age'])
plt.title('Box plot of Doors')
plt.ylabel('Doors')
cars_df.describe()
```

```python
sns.boxplot(x=cars_df['FuelType'],y=cars_df['Price'],palette="Set1")
plt.title('Box plot of Price')
plt.xlabel('Fuel Type')
```

# **Histogram plot**

```python
sns.histplot(data=cars_df['Price'],bins=20)
```

```python
sns.countplot(x=cars_df['FuelType'])
plt.title('Frequency distribution of Fuel Type')
```

# **Scatter Plot**

```python
sns.set(style='darkgrid')
sns.regplot(x=cars_df['Age'],y=cars_df['Price'],fit_reg=True,marker="*")
plt.title('Age vs Price')
plt.xlabel('Age')
plt.ylabel('Price')
```

```python
plt.scatter(cars_df['KM'],cars_df['Price'])
```

# Multivariable Scatter plot

```python
sns.lmplot(x='Age',y='Price',data=cars_df,fit_reg=False,hue='FuelType',legend='True',palette="Set1",markers="*")
plt.title('Age vs Price')

plt.xlabel('Age')
plt.ylabel('Price')
```

# **Pairwise plot**
