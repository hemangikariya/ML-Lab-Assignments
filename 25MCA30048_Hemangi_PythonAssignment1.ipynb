# **Part B: Working with Lists and Dictionaries**

```python
# Nested List Access
lst = [1,2,[3,4],[5,[100,200,["hello"]],23,11],1,7]
print(lst[3][1][2][0])
```

```python
# Nested Dictionary Access
d = {'k1':[1,2,3,{'tricky':['oh','man','inception',{'target':[1,2,3,'hello']}]}]}
print(d['k1'][3]['tricky'][3]['target'][3])
```

```python
# Theory Question: Difference Between List and Tuple
# List:
#- It is mutable (can be changed)
#- It uses square brackets []
#- More memory usage
#- Methods like append(), remove(), available
my_list = [1, 2, 3]
my_list.append(4)
print(my_list)

# Tuple
#- It is immutable (cannet be changed)
#- It uses parentheses ()
#- Less memory usage
#- Limited methods available
my_tuple = (1, 2, 3)
print(my_tuple)

# Difference
# Lists are mutable; tuples are immutable.
# Lists use []; tuples use ().
# Lists consume more memory and are slower than tuples.
```
