```python
# Theory Question: Difference Between List and Tuple
# List:
#- It is mutable (can be changed)
#- It uses square brackets []
#- More memory usage
#- Methods like append(), remove(), available
my_list = [1, 2, 3]
my_list.append(4)
print(my_list)

# Tuple
#- It is immutable (cannet be changed)
#- It uses parentheses ()
#- Less memory usage
#- Limited methods available
my_tuple = (1, 2, 3)
print(my_tuple)

# Difference
# Lists are mutable; tuples are immutable.
# Lists use []; tuples use ().
# Lists consume more memory and are slower than tuples.
```

# **Part C: User Defined Functions**

```python
# Email Domain Extracter
def domainGet(email):
  return email.split('@')[1]

print(domainGet('user@gmail.com'))
```

```python
# Dog Finder
import string

def findDog(sentence):
  sentence = sentence.lower()
  for p in string.punctuation:
    sentence = sentence.replace(p, "")
    return "dog" in sentence.split()
print(findDog("Is there a Dog here?"))
```

```python
# Dog Counter
import string

def countDog(sentence):
    sentence = sentence.lower()

    for p in string.punctuation:
        sentence = sentence.replace(p, "")

    return sentence.split().count("dog")

print(countDog("This dog runs faster than the other dog."))
```

**Part D: Conditional Statements**

```python
# Speeding Ticket Calculator
def caught_speeding(speed, is_birthday):

    if is_birthday:
        speed -= 5

    if speed <= 60:
        return "No Ticket"
    elif speed <= 80:
        return "Small Ticket"
    else:
        return "Big Ticket"


print(caught_speeding(81, True))
print(caught_speeding(81, False))
```

**Part E: Additional Programming Exercise**

```python
# Menu-Driven Application
import string

def domainGet(email):
    return email.split('@')[1]

def findDog(sentence):
    sentence = sentence.lower()

    for p in string.punctuation:
        sentence = sentence.replace(p, "")

    return "dog" in sentence.split()

def countDog(sentence):
    sentence = sentence.lower()

    for p in string.punctuation:
        sentence = sentence.replace(p, "")

    return sentence.split().count("dog")

while True:

    print("\n----- MENU -----")
    print("1. Extract Email Domain")
    print("2. Find Word 'dog'")
    print("3. Count Occurrences of 'dog'")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        email = input("Enter email: ")
        print("Domain:", domainGet(email))

    elif choice == "2":
        sentence = input("Enter sentence: ")
        print(findDog(sentence))

    elif choice == "3":
        sentence = input("Enter sentence: ")
        print("Count:", countDog(sentence))

    elif choice == "4":
        print("Exiting program...")
        break

    else:
        print("Invalid choice!")
```
